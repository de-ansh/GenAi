```python
from dotenv import load_dotenv
load_dotenv()
import os
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
HUGGINGFACE_API_KEY = os.getenv("HUGGINGFACE_API_KEY")

```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model_name="gpt-4", temperature=0)


```

```python
res=llm.invoke("Hello, my name is Ashish. Can you help me write a poem about me?")
print(res.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm_google = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
response = llm_google.invoke("Hello, my name is Ashish. WHats Main difference between Gemini 1.5 and Gemini 2.5?")
print(response.content)
```
